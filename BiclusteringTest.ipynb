#Biclustering algorithm: δ-biclustering (Cheng & Church or CC)

```python
%%latex
Define Mean Square Residue (MSR) $H(I,J)$:
$$H(I,J) = \frac{1}{|I||J|} \sum_{i \in I} \sum_{j \in J} (a_{ij} - a_{Ij} - a_{iJ} + a_{IJ})^2$$
where:
$$a_{iJ} = \frac{1}{|J|} \sum_{j \in J} a_{ij}$$
$$a_{Ij} = \frac{1}{|I|} \sum_{i \in I} a_{ij}$$
$$a_{IJ} = \frac{1}{|I||J|} \sum_{i \in I, j \in J} a_{ij}$$
```

```python
class MSR:
    def __init__(self, data):
        n,m = (len(data), len(data[0]))
        self.aiJ = 1.0/m * np.sum(data, axis=0)
        self.aIj = 1.0/n * np.sum(data, axis=1)
        self.aIJ = 1.0/(n*m) * np.sum(data)

        # it could be changed to change the shape of aiJ and aIj to simplify the code
        #   by calculating H in a vectorized way, but it'd spend more memory
        self.H = 0
        self.HiJ = np.zeros(m)
        self.HIj = np.zeros(n)
        for i in xrange(n):
            for j in xrange(m):
                self.H += ( data[i,j] - self.aIj[i] - self.aiJ[j] + self.aIJ )**2
                self.HiJ[j] += ( data[i,j] - self.aIj[i] - self.aiJ[j] + self.aIJ )**2
        self.H *= 1.0/(n*m)
        self.HIj *= 1.0/n

        for j in xrange(m):
            for i in xrange(n):
                self.HIj[i] += ( data[i,j] - self.aIj[i] - self.aiJ[j] + self.aIJ )**2
        self.HiJ *= 1.0/m


msr = MSR(data_shuffled)
print msr.H
print msr.HiJ
print msr.HIj
```

```python
def remove_unique_nodes(data, delta, I=None, J=None):
    if not I:
        I = np.arange(len(data))
    if not J:
        J = np.arange(len(data[0]))
    
    while (msr.H > delta):
        msr = MSR(data)

        lines_idxs_to_remove = np.argmax(msr.HIj)
        cols_idxs_to_remove = np.argmax(msr.HiJ)
        if msr.HIj[lines_idxs_to_remove] > msr.HiJ[cols_idxs_to_remove]:
            I = np.remove(I, lines_idxs_to_remove)
            data = data[I]
        else:
            J = np.remove(J, cols_idxs_to_remove)
            data = data[:,J]
    return (data,I,J)
```
