#Biclustering algorithm: δ-biclustering (Cheng & Church or CC)

```python
%%latex
Define Mean Square Residue (MSR) $H(I,J)$:
$$H(I,J) = \frac{1}{|I||J|} \sum_{i \in I} \sum_{j \in J} (a_{ij} - a_{Ij} - a_{iJ} + a_{IJ})^2$$
where:
$$a_{iJ} = \frac{1}{|J|} \sum_{j \in J} a_{ij}$$
$$a_{Ij} = \frac{1}{|I|} \sum_{i \in I} a_{ij}$$
$$a_{IJ} = \frac{1}{|I||J|} \sum_{i \in I, j \in J} a_{ij}$$
```

```python
class MSR:
    def __init__(self, data, shape):
        I,J = shape
        self.aiJ = 1.0/J * np.sum(data, axis=0)
        self.aIj = 1.0/I * np.sum(data, axis=1)
        self.aIJ = 1.0/(I*J) * np.sum(data)
        print len(self.aiJ)
        self.H = 1.0/(I*J) * np.sum(( data - self.aIj - self.aiJ + self.aIJ )**2)
        self.HJ = 1.0/J * np.sum(( data - self.aIj - self.aiJ + self.aIJ )**2, axis=0)
        self.HI = 1.0/I * np.sum(( data - self.aIj - self.aiJ + self.aIJ )**2, axis=1)


msr = MSR(data_shuffled,shape)
print msr.H
print len(msr.HJ)
print len(msr.HI)
```

```python
def remove_unique_node(delta, data, shape):
    n,m = shape
    L = np.arange(n)
    J = np.arange(m)
    
    while(msr.H > delta):
        msr = MSR(data,shape)
        
        line_index_to_remove = np.argmax(msr.aiJ)
        col_index_to_remove = np.argmax(msr.aIj)
        if msr.aiJ[line_index_to_remove] > msr.aIj[col_index_to_remove]:
            L = np.remove(L, line_index_to_remove)
            data = data[L]
        else:
            J = np.remove(J, col_index_to_remove)
            data = data[J]
```
