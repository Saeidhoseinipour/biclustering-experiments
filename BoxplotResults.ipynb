```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style="ticks", palette="muted", color_codes=True)
%matplotlib inline
# Load the example planets dataset
planets = sns.load_dataset("planets")

# Plot the orbital period with horizontal boxes
ax = sns.boxplot(x="distance", y="method", data=planets,
                 whis=np.inf, color="c")

# Add in points to show each observation
sns.stripplot(x="distance", y="method", data=planets,
              jitter=True, size=3, color=".3", linewidth=0)
plt.show()
```

```python
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
plt.rcParams['figure.figsize'] = (25.0, 5.0)

plt.rcParams['font.family'] = 'normal'
# plt.rcParams['font.weight'] = 'bold'
plt.rcParams['font.size'] = 40
sns.set(font_scale=1.3)
sns.set_style("whitegrid")
```

```python
files = {
    'k-means': '/Users/lucasbrunialti/git/biclustering/experiments/ig_kmeans_news_results.csv',
    'fuzzy k-means': '/Users/lucasbrunialti/git/biclustering/experiments/ig_fuzzy_cmeans_news_results.csv',
    'ONMTF': '/Users/lucasbrunialti/git/biclustering/experiments/backup/onmtf/ig_onmtf_news_result.csv',
    'FNMTF': '/Users/lucasbrunialti/git/biclustering/experiments/backup/fnmtf/ig_fnmtf_news_result.csv',
    'OvNMTF': '/Users/lucasbrunialti/git/biclustering/experiments/backup/ovnmtf/ig_ovnmtf_news_result.csv',
    'BinOvNMTF': '/Users/lucasbrunialti/git/biclustering/experiments/backup/bin_ovnmtf/ig_bin_ovnmtf_news_results.csv'
}

results_dfs = []
for algo, filename in files.iteritems():
    df = pd.read_csv(filename)
    df['algo'] = algo
    if algo == 'k-means' or algo == 'fuzzy k-means':
        df['L'] = df['K']
    results_dfs.append(df)
results_df = pd.concat(results_dfs)
```
